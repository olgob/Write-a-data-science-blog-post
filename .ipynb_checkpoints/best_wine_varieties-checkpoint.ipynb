## Import libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Load data

```python
df = pd.read_csv("winemag-data_first150k.csv") 
del df['Unnamed: 0']
df.head()
```

## What are the best producers of wine?

We know that 5 values are missing from the column "country". What about the column "points"

```python
df['points'].isnull().sum()
```

No values are missing. We just have to supress the rows where country is not defined

```python
df = df.dropna(subset=['country'])
```

We check that the rows have been supressed

```python
df['country'].isnull().sum()
```

Now let´s group the data by country and let´s see the average rate for each.

```python
country_rate_avg = df.groupby('country')['points'].mean().sort_values(ascending=False)
plt.rcParams["figure.figsize"] = (20,10)
fig, ax = plt.subplots()
ax.bar(country_rate_avg.index, country_rate_avg.values)
plt.setp(ax.get_xticklabels(), rotation=90);# country_wines_n/df.shape[0]
ax.set_ylabel('Average rate')
ax.set_ylim([81, 94])
```

Even if England is one of the smallest provider of wine, it produces on the average the best wine. But it is maybe not really representative...

## What if we look the distribution of the points for each country?

We create a new columns which will serve to count the wine with an excellent grade (>=95)

```python
df_select = df[(df['country']=='US') |
               (df['country']=='Italy') | 
               (df['country']=='France') | 
               (df['country']=='Spain') |
               (df['country']=='Argentina') |
               (df['country']=='Portugal') |
               (df['country']=='Australia') |
               (df['country']=='New Zealand') |
               (df['country']=='Austria') |
               (df['country']=='Germany')]
```

```python
sns.violinplot(x=df_select['country'], y=df_select['points'], data=df_select, inner=None);
```

Among the top 10 biggest producers, France has the bigger proportion of wines above 95.

## Conclusion

France has the highest proportion of wine that are rated above 95 by far followed by US, Germany, Austria, and Italy.
